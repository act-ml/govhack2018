```python
import csv
```

```python
filename = 'grants_mk2.csv'
filename_out = 'grants_mk3.csv'

filename_mk1 = 'grants.csv'

header = ''

with open(filename_mk1, 'rb') as csvfile:
    grants = csv.reader(csvfile, delimiter=',', quotechar='"')
    for row in grants:
        header = row[:-1]
        break
        
```

```python
filename_coord = 'gnaf/australian_postcodes.csv'
location = dict()

with open(filename_coord, 'rb') as in_f:
    loc = csv.reader(in_f, delimiter=',', quotechar='"')
    
    for i, row in enumerate(loc):
        if i == 0: continue
            
        #print row
        postcode, _, _, lon, lat, _, _ = row
        location[postcode] = [lat, lon]
        
```

```python
# 30 app sub
# 32 app postcode
# 33 app st

# P 15, Purpose

grants_text = dict() 

with open(filename, 'rb') as csvfile:
    with open(filename_out, 'w') as out_f:
        #grants = csv.reader(csvfile, delimiter=' ', quotechar='|')
        grants = csv.reader(csvfile, delimiter='\t', quotechar='"')

        for i, row in enumerate(grants):
            if i == 0: out_f.write( '\t'.join( [str(r) for r in header] + ['lat', 'lon'] ) + '\n') 
            #if i > 100: break
            #if len(row) < 30: continue
            if len(row) < 30:
                out_f.write( '\t'.join( [str(r) for r in row] + ['', ''] ) + '\n') 
                continue

            #print row
            app_sub = row[30]
            
            try:
                app_postcode = int(float( row[32] ))
            except ValueError:
                app_postcode = 0
                
            app_st = row[33]
            
            try:
               del_postcode = int(float( row[36] ))
            except ValueError:
               del_postcode = app_postcode
               row[36] = app_postcode
                
            try: 
                lat, lon = location[str(del_postcode)]
            except KeyError:
                lat = ''
                lon = ''
            
            row[32] = 1000
            
            #if del_postcode
            #print app_sub, app_postcode, app_st, int(float(del_postcode) )
            #print row
            out_f.write( '\t'.join( [str(r) for r in row[:-1]] + [str(lat), str(lon)] ) + '\n') 
            
        
```
